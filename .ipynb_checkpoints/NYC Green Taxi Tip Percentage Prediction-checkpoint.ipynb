# Feature Engineering

# Let's do some Feature Engineering
The assumption here is that the hour of the day and the day of the month can have a bearing with the tip percentage, hence these two derived fields were included in the final dataset.
Another assumption is about the varibale trip duration. Trip duration can influence the tip percentage. It might very well happen that a shorter trip duration has positive corelation with larger tip or vice-versa. Hence, we include this derived field as well into our data set

```python
# Deriving Hour Of the Day
adata['Hour_Of_The_Day'] = pd.to_datetime(adata['lpep_pickup_datetime']).dt.hour
```

```python
#Deriving Day Of The Month
adata['Day_Of_The_Month'] = pd.to_datetime(adata['lpep_pickup_datetime']).dt.day
```

```python
# Deriving Trip Duration
adata['Trip_Duration'] = pd.to_datetime(adata['Lpep_dropoff_datetime'])-pd.to_datetime(adata['lpep_pickup_datetime'])
```

```python
# Converting trip duration to seconds
adata['Trip_Duration']= adata['Trip_Duration'].dt.seconds
```

In couple of the below steps, we are going to seperate train and test data

```python
#Seperate the target variable tip_percent into in a data frame
y = pd.DataFrame(adata['tip_percent'])
```

```python
#Seperate the feature variables into a data frame
X = adata.drop(columns = 'tip_percent')
```

```python
# Splitting the feature and target variables into train and test sets
from sklearn.model_selection import train_test_split
feature_df, feature_df_test, target_df, target_df_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# creating a combined data set by concatenating feature and target train sets
raw_train_df = pd.concat([feature_df,target_df], axis =1 )
```

```python
raw_train_df.columns
```

```python
feature_df.columns
```

```python
# Seperating the categorical and numerical variables in seperate lists. This would be useful while doing one hot encoding
categorical_vars = ['Store_and_fwd_flag','RateCodeID','Payment_type','Trip_type']
numeric_vars = ['Pickup_longitude','Pickup_latitude','Dropoff_longitude',
                'Dropoff_latitude','Passenger_count','Trip_distance','Fare_amount','Extra','MTA_tax','Tolls_amount',
                'improvement_surcharge','Total_amount','Hour_Of_The_Day','Day_Of_The_Month','Trip_Duration']
```

```python
# Defining functions that will be required at the time of model building

def one_hot_encode_feature_df(df, cat_vars=None, num_vars=None):
    '''performs one-hot encoding on all categorical variables and combines result with continous variables'''
    cat_df = pd.get_dummies(df[cat_vars])
    num_df = df[num_vars].apply(pd.to_numeric)
    return pd.concat([cat_df, num_df], axis=1)#,ignore_index=False)

# function for doing cross validation using the model passed to it
def train_model(model, feature_df, target_df, num_procs, mean_mse, cv_std):
    neg_mse = cross_val_score(model, feature_df, target_df, cv=5, n_jobs=num_procs, scoring='neg_mean_squared_error')
    mean_mse[model] = -1.0*np.mean(neg_mse)
    cv_std[model] = np.std(neg_mse)
    
def print_summary(model, mean_mse, cv_std):
    print('\nModel:\n', model)
    print('Average MSE:\n', mean_mse[model])
    print('Standard deviation during CV:\n', cv_std[model])
    
def save_results(model, mean_mse, predictions, feature_importances):
    '''saves model, model summary, feature importances, and predictions'''
    with open('model.txt', 'w') as file:
        file.write(str(model))
    feature_importances.to_csv('feature_importances.csv') 
    np.savetxt('predictions.csv', predictions, delimiter=',')
```

```python
#encode categorical data and get final feature dfs
print("Encoding data")
feature_df = one_hot_encode_feature_df(raw_train_df, cat_vars=categorical_vars, num_vars=numeric_vars)
test_df = one_hot_encode_feature_df(feature_df_test, cat_vars=categorical_vars, num_vars=numeric_vars)
```

```python
feature_df.shape
```

```python
test_df.shape
```

In couple of steps below we are going to build the model.

This is a regression problem as we are predicting the tip percentage here, but from the exploratory analysis done above we discovered that the relation of independent and dependent variables are not linear. Hence we refrain from using linear regresion models and would rather choose ensemble regression models Random Forest Regressor and Gradient Boosting.

I chose the mean squared error (MSE) over R-squared as the evaluation metrics as it gives a better idea about how far off the predicted values are from the actual values.

```python
#initialize model list and dicts
models = []
mean_mse = {}
cv_std = {}
res = {}

#define number of processes to run in parallel
num_procs = 4

#shared model paramaters
verbose_lvl = 0
```

```python
#create models -- hyperparameter are taken as the best guess and based on previous experience.

rf = RandomForestRegressor(n_estimators=150, n_jobs=num_procs, max_depth=25, min_samples_split=60, \
                           max_features=25, verbose=verbose_lvl)
gbm = GradientBoostingRegressor(n_estimators=150, max_depth=5, loss='ls', verbose=verbose_lvl)
                      
models.extend([rf, gbm])

#parallel cross-validate models, using MSE as evaluation metric, and print summaries
print("Beginning cross validation")
for model in models:
    train_model(model, feature_df, target_df, num_procs, mean_mse, cv_std)
    print_summary(model, mean_mse, cv_std)
```

As mentioned above, the hyperparameters were chosen as the best guess and based on previous experience. If i had sifficient time i would have tried grid search to tune the model further. I would have also tried to transform the independent variables and run the regression.

The MSE for GBM (1.466) is better than random forest(1.667), with a lower standard deviation and hence I will choose that as my final model to predict the tip_percent in the test data.The MSE appears to be good and gives us confidence that model will predict the unseen data well.

```python
#choose model with lowest mse
model = min(mean_mse, key=mean_mse.get)
print('\nPredictions calculated using model with lowest MSE:')
print(model)

#train model on entire dataset
model.fit(feature_df, target_df)
#create predictions based on test data
predictions = model.predict(test_df)
```

```python
predictions
```

```python
# calculate the residual from actual values and the predicted values in the test data set
resid = np.array(target_df_test['tip_percent'])- predictions
```

```python
# Plot the distribution of residuals
pd.DataFrame(resid).hist(bins=20)
plt.yscale('log')
plt.xlabel('predicted - real')
plt.ylabel('count')
plt.title('Residual plot')
plt.show()
```

The residual plot is roughly symetrical, which shows an equal bias.

```python
#calculate the MSE using the actual and predicted value
from sklearn import metrics
print ('RFR test mse:',metrics.mean_squared_error(predictions,np.array(target_df_test['tip_percent'])))
print ('RFR r2:', metrics.r2_score(predictions,np.array(target_df_test.tip_percent)))
```

When the model was applied on the unseen test data it performed seemingly well and returned a MSE that is quite similar to the MSE calculated on the train data. It also returns a R-squared value of .9784 which means that the model rougly explains 98% of the variance in the target variable.
