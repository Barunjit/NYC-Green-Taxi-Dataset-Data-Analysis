## Questio:
Plot a histogram of the number of the trip distance (“Trip Distance”).

```python
#Plotting a histogram for the column Trip_distance to have a look at the frequency distribution
%matplotlib inline
plt.figure( figsize = (10,5))
plt.hist(data['Trip_distance'], color='blue')
plt.xlabel('Trip Distance')
plt.ylabel('# of Trip Distance')
plt.title('Histogram of Trip Distance',fontsize = 15)
```

**Observation** : The above histogram clearly shows that most of the trip distances fall in the bracket of 0-50 miles. But to have a closer look let us plot histogram for each of the 100 miles range. 

```python
plt.subplots(2,3,figsize = (20,10))
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.hist(data['Trip_distance'],bins = 10,range = [i*100,(i+1)*100], color='blue')
```

**Observation**: The above histograms further shows that the majority of trip durations are in the range 0-20, more so in the range 0-10 miles. To have a more detailed look let's drill down to every 5 miles range in the overall range of 0-20 miles as that's where most of the trips have taken place.

```python
plt.subplots(2,2,figsize = (20,10))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.hist(data['Trip_distance'],bins = 10,range = [i*5,(i+1)*5], color='blue')
```

**observation**: The above histograms further narrow down the range to 0-5 miles, range where maximum trips took 
place. Let's go a little bit deeper to find out in the 0-5 miles range,which subrange of trip duration has most counts.

```python
plt.subplots(2,3,figsize = (20,10))
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.hist(data['Trip_distance'],bins = 10,range = [i,(i+1)], color='blue')
```

**observation** : The above graphs clearly show that in the ovreall 0-5 range most of the trip distance are in the range 0-1 and 1-2 miles. Infact the most number of trip_distances across all the ranges is in the range 0.9-1.0 miles. This number is around 70000.

## Question 3(a):
Report mean and median trip distance grouped by hour of day.

```python
meanbyhour = bdata.groupby(pd.to_datetime(data['lpep_pickup_datetime']).dt.hour)['Trip_distance'].mean()
meanbyhour
```

```python
medianbyhour = bdata.groupby(pd.to_datetime(data['lpep_pickup_datetime']).dt.hour)['Trip_distance'].median()
medianbyhour
```

The trip distances were grouped by hour of the day and then the mean and medians were calulated and displayed above.

## Let us analyse it a little more

```python
plt.figure(figsize=(10,5))
plt.plot(meanbyhour, color='red',label='Mean by Hour')
plt.plot(medianbyhour, color = 'blue',label = 'Median By hour')
plt.xticks(np.arange(0,24, step=1))
plt.legend()
plt.xlabel('Hour Of The Day')
plt.ylabel('Mean Trip Distance')
plt.title('Mean and Median Trip Distance Over The Day')
```

After plotting the mean and median trip distances over the day we could see that there is a peak in the early hours (4-6) and then it falls sharply,and then it again starts peaking after 6 PM, achieving a peak around 10 PM. I would assume that people try to avoid the traffic by commuting in early mornings or after the office hours, i.e. after 6 PM. The (4-6) peak is higher than the peak around 10 PM. I would assume this is because most people who work would prefer to avoid traffic in the morning to be at work on time.

## Question 3(b):
We’d like to get a rough sense of identifying trips that originate or terminate at one of the NYC area airports.
Can you provide a count of how many transactions fit this criteria, the average fare, and any other interesting
characteristics of these trips.

The data dictionary mentions the RateCodeID for JFK and Newark as 2 and 3 respectively. The assumption here is that the trips having these ratecodes are the trips that originate or terminate at one of the NYC areas airports. I would have liked to use the pickup and dropoff latitudes and longitides to more precisely identify these trips, but because of the time constraint I have preferred to stick to this method.  

```python
bdata = data.copy()
```

```python
# Creating a new column to find out the hour of the day for each trip
bdata['Hour_Of_The_Day'] = pd.to_datetime(bdata['lpep_pickup_datetime']).dt.hour
```

```python
# Using RateCodeID Seggregating the trip that originates/terminates at NYC aiports versus the ones that do not.
airtrip = bdata[bdata['RateCodeID'].isin([2,3])]
noairtrip = bdata[~bdata['RateCodeID'].isin([2,3])]
```

The number of trips that originates/terminates at one of the NYC area airports

```python
print('The Number of trips originating or terminationg at one of the NYC areaairports is '+ str(airtrip.shape[0]))
```

The average fare of trips that originates/terminates at one of the NYC area airports

```python
print('The average fare amount for airport trips is '+ str(airtrip['Fare_amount'].mean()))
#print(noairtrip['Fare_amount'].mean())
```

Let us compare the average fare amount for the airport and nonairport trips

```python
print('The average Fare amount for airport trips is '+ str(airtrip['Fare_amount'].mean()))
print('The average Fare amount for nonairport trips is '+ str(noairtrip['Fare_amount'].mean()))
```

This will be interesting!! Let us compare the average tip amount for the airport and nonairport trips

```python
print('The average tip amount for airport trips is '+ str(airtrip['Tip_amount'].mean()))
print('The average tip amount for nonairport trips is '+ str(noairtrip['Tip_amount'].mean()))
```

Let us also calculate the average total amount for the airport and nonairport trips

```python
print(airtrip['Total_amount'].mean())
print(noairtrip['Total_amount'].mean())
```

Here I intend to compare the trend of mean tip amount for airport trips versus nonairport trips over the day. Let's plot the corresponding line plot.

```python
t1 = airtrip.groupby('Hour_Of_The_Day')['Tip_amount'].mean()
t2 = noairtrip.groupby('Hour_Of_The_Day')['Tip_amount'].mean()
```

```python
plt.figure(figsize=(10,5))
plt.plot(t2, color='red',label='non_airport_trips')
plt.plot(t1, color = 'blue',label = 'airport_trips')
plt.xticks(np.arange(0,24, step=1))
plt.legend()
plt.xlabel('Hour Of The Day')
plt.ylabel('Mean Tip Amount')
plt.suptitle('Tip Amount trend for airport versus non airport trips over the day')
```

**Observation** : This is interesting!! The tip amount for the airport trips increases steeply in between 2-4 PM and again peaks between 1:00-2:00 AM. The tip for non airport trips on the other hand remains more or less same with a small increase between 4:00-6:00 AM. 

Here I intend to compare the trend of number of trips for airport trips versus nonairport trips over the day. Let's plot the corresponding line plot.

```python
c1 = airtrip.groupby('Hour_Of_The_Day')['Trip_distance'].count()
c2 = noairtrip.groupby('Hour_Of_The_Day')['Trip_distance'].count()
```

```python
plt.figure(figsize=(20,20))
plt.subplots(1,2, figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(c1,color='red',label='airport_trips')
plt.xticks(np.arange(0,24, step=1))
plt.legend()
plt.xlabel('Hour Of The Day')
plt.ylabel('Number Of Trips')

plt.subplot(1,2,2)
plt.plot(c2,color = 'blue',label = 'non_airport_trips')
plt.xticks(np.arange(0,24, step=1))
plt.legend()
plt.xlabel('Hour Of The Day')
plt.ylabel('Number Of Trips')
plt.suptitle('# of Trips for Airport Versus NonAirport trips over the day')
```

**Observation**: The number of aiport trips gradually increases from 02:00 AM onwards peaks in between 01:00-03:00 PM and decreases then onwards. Whereas the nonairport trips decreases 12:00 AM onwards before it starts increasing after 05:00 AM. It flattens for a duration between 08:00 AM - 01:00 PM before again increasing and hitting a peak at around 06:00 PM.
