
### Building a derived variable representing the average speed over the course of a trip and performing statistical analysis to conclude whether there is any substantial difference between the average 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
#download and load the data into a pandas dataframe
data = pd.read_csv('https://s3.amazonaws.com/nyc-tlc/trip+data/green_tripdata_2015-09.csv')
```

```python
#Making a copy of data so that if something goes wrong we don't have to download the data again
bdata = data.copy()
```

```python
# Finding out the count and data types of the columns
bdata.info()
```

```python
# Calculate the Trip duration using pickup and drop times 
bdata['Trip_Duration'] = pd.to_datetime(bdata['Lpep_dropoff_datetime'])-pd.to_datetime(bdata['lpep_pickup_datetime'])
```

```python
# Converting the Trip duratuion in seconds
bdata['Trip_Duration']= bdata['Trip_Duration'].dt.seconds
```

```python
# Drawing a histogram of the trip durations
%matplotlib inline
plt.figure( figsize = (10,5))
plt.hist(bdata['Trip_distance'], color='blue')
plt.xlabel('Trip Distance')
plt.ylabel('# of Trip Distance')
plt.title('Histogram of Trip Distance',fontsize = 15)
#bdata['Trip_Duration'].hist()
```

```python
# Describing the distribution of Trip Duration
bdata['Trip_Duration'].describe()
```

```python
bdata[bdata['Trip_Duration']< 20].shape
```

Here the assumption is that the trips less than 20 seconds are not valid and they might have been recorded wrongly in the dataset. As the average is very sensitive to outliers, to have a more accurate result we will discard there records from our data set before analysing the data. In conclusion, we will only consider the trips with trip duration greater than 20 seconds 

```python
bdata = bdata[bdata['Trip_Duration']> 20]
```


(b) Build a derived variable representing the average speed over the course of a trip.Can you perform a test to determine if the average trip speeds are materially the same in all weeks of
September? If you decide they are not the same, can you form a hypothesis regarding why they differ?

```python
# Building a derived variable using the Trip_distance and Trip Duration
bdata['average_speed'] = 3600*(bdata['Trip_distance']/bdata['Trip_Duration'])
```

```python
bdata.info()
```

```python
# Deriving the week of trip of each trip
bdata['week'] = pd.to_datetime(bdata['lpep_pickup_datetime']).dt.week
```

The above derivation gives us the week for each trip. In the below output we can see that there are five weeks available is september data and the week numbers are nothing but week number of the year (36,37,38,39,40)

```python
# calcuating the number of trips for each week
bdata['week'].value_counts()
```

Here we segregate and store entries of average speed for each week and also print there respective counts

```python
week1_average = np.array(bdata[bdata['week']==36]['average_speed'])
week2_average = np.array(bdata[bdata['week']==37]['average_speed'])
week3_average = np.array(bdata[bdata['week']==38]['average_speed'])
week4_average = np.array(bdata[bdata['week']==39]['average_speed'])
week5_average = np.array(bdata[bdata['week']==40]['average_speed'])
```

```python
print(len(week1_average))
print(len(week2_average))
print(len(week3_average))
print(len(week4_average))
print(len(week5_average))
```

```python
# Calculating the tmeans of average speeds by week and storing in a list
week_list= [week1_average.mean(), week2_average.mean(), week3_average.mean(),week4_average.mean(),week5_average.mean()]
```

```python
week_list
```

```python
index = np.arange(len(week_list))
```

```python
# Plotting the means of average speeds by week
plt.figure(figsize=(10,5))
sns.set(style="dark", font_scale=1)
sns.barplot(index, week_list)
plt.title('Mean Of Average speeds by Week', fontsize=15)
plt.xlabel('Week Of The Month', fontsize = 15)
plt.ylabel('Mean of Average Speeds', fontsize = 15)
```

From the above bar graph we can visually see some difference in the means of average speeds across the weeks, but whether these diffrences are statistically significant? Let's perform an ANOVA test and have a look at the p-value. The p-value is almost zero, so assuming a significance level of 0.05, we reject the null that all the weekly means of average speeds are same and conclude that atleast two of them are statistically different. The reason for this diffrence has to be further investigated. In conclusion, the average trip speeds across the weeks are materially diffrent.

```python
stats.f_oneway(week1_average,week2_average, week3_average,week4_average, week5_average)
```

```python
print(week1_average.mean(),week2_average.mean(),week3_average.mean(),week4_average.mean(),week5_average.mean())
```


(c) Can you build up a hypothesis of average trip speed as a function of time of day?

```python
# Creating a column in the dataframe that stores the hour of the day for each trip
bdata['Hour_Of_The_Day'] = pd.to_datetime(bdata['lpep_pickup_datetime']).dt.hour
```

```python
bdata['Hour_Of_The_Day'].dtype
```

```python
# Segregating and storing the average speeds for each hour in an array 
avgbyhour = []
for i in range(0,24):
    avgbyhour.append(bdata[bdata['Hour_Of_The_Day']== i]['average_speed'])
```

```python
# Performing an anova test to conclude whether average speeds statistically differ across hour of the day
stats.f_oneway(avgbyhour[0],avgbyhour[1],avgbyhour[2],avgbyhour[3],avgbyhour[4],
               avgbyhour[5],avgbyhour[6],avgbyhour[7],avgbyhour[8],avgbyhour[9],
               avgbyhour[10],avgbyhour[11],avgbyhour[12],avgbyhour[13],
               avgbyhour[14],avgbyhour[15],avgbyhour[16],avgbyhour[17],avgbyhour[18],
               avgbyhour[19],avgbyhour[20],avgbyhour[21],avgbyhour[22],avgbyhour[23])
```

As in case of average speeds across weeks, The anova test here also returns a p value that is very less (zero) than the significance level of 0.05 and hence we reject the null and conclude that the average speeds are statistically diffrenet across the different hours of the day.

```python
#calculating the mean of average speeds across hour of the day
meanbyhour = []
medianbyhour = []
for i in range(0,24):
    meanbyhour.append(avgbyhour[i].mean())
    medianbyhour.append(avgbyhour[i].median())   
```

```python
index1 = np.arange(len(meanbyhour))
```

```python
#Plotting the mean of average speeds across hour of the day
plt.figure(figsize=(10,5))
sns.set(style="dark", font_scale=1)
sns.barplot(index1, meanbyhour, color = 'cyan')
plt.title('Mean Of Average Speeds By Hour', fontsize=15)
plt.xlabel('Hour OF The Day', fontsize = 15)
plt.ylabel('Mean Of Average Speeds', fontsize = 15)
```

```python
plt.figure(figsize=(10,5))
sns.set(style="dark", font_scale=1)
sns.barplot(index1, medianbyhour, color = 'blue')
plt.title('Median Of Average speed by hour', fontsize=15)
plt.xlabel('Hour OF The Day', fontsize = 15)
plt.ylabel('Median Of Average Speed', fontsize = 15)
```

**Observation**: Both the above graphs of mean and meadin average speeds across hour of the day shows a common trend. The average speeds keep increasing from 12:00 AM onwards and peaks at 05:00-06:00 AM before coming down. It Again start increasing after 05:00 PM. The reason behind this may be the empty roads/less traffic at these times that encourages the drivers to drive fast.
